```python
import pandas as pd
import numpy as np

import folium
```

```python
df_xfit=pd.read_csv('data_file.csv')
df_xfit_sub=df_xfit[['country','city','name','active','latlon']]
df_xfit_sub=df_xfit_sub.dropna()
df_france=df_xfit_sub[df_xfit_sub['country']=='France']
df_france=df_france.iloc[0:50,:]
df_france[['lat', 'lon']] = df_france.latlon.str.split(',' ,expand=True)
df_france[['lat', 'lon']]=df_france[['lat', 'lon']].astype(float)
```

```python
worldmap=folium.Map(location=[48.862725,2.287592],zoom_start=8)

for lat, lng, label in zip(df_france.lat,df_france.lon,df_france.name):
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    color='yellow',
    fill=True,
    popup=label,
    fill_color='blue',
    fill_opacity=0.6).add_to(worldmap)
    
worldmap
```
